```python
import statsmodels.regression.linear_model as sm
X=np.append(arr=np.ones( (50,1)).astype(int),values=X,axis=1)
X
```

```python
#backward elimination
X_opt= X[:,[0,1,2,3,4,5]]
reg_ols=sm.OLS(endog=Y,exog=X_opt).fit()
```
